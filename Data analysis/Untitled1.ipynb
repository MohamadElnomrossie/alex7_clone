```python
import pandas as pd
import statsmodels.api as sm
from matplotlib import pyplot as plt
import seaborn as sb
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("house_price_area_only.csv")
```

```python
df=df[['price','area']]
```

```python
df.head()
```

```python
sb.scatterplot(x='area',y='price',data=df);
```

```python
df['intercept']=1
```

```python
df.head()
```

# Ordinary Least Squares (OLS)

```python
lm=sm.OLS(df['price'],df[['area','intercept']])
res=lm.fit()
```

```python
res.summary()
```

```python
lr=LinearRegression()
```

```python
lr.fit(df['area'].to_numpy().reshape(-1,1),
       df['price'].to_numpy().reshape(-1,1))
```

```python
lr.intercept_
```
