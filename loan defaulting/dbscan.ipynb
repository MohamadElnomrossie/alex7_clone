```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
pd.set_option('display.max_columns',None)
```

```python
df=pd.read_csv("train.csv")
```

# Data Understanding

```python
df.head()
```

```python
df.info()
```

```python
num_data=cat_data=df.select_dtypes('number')
num_data.drop(columns=['Loan Status'],inplace=True)
```

```python

scaler=MinMaxScaler()
df[num_data.columns]=scaler.fit_transform(df[num_data.columns])
```

```python
cat_data=df.select_dtypes('object')
```

```python
for col in cat_data.columns:
    print(col)
    print(cat_data[col].nunique())
    print('===============================')
```

```python
# Data Cleaning
df.drop(columns=['Payment Plan'],inplace=True)
```

# Turning object into numerical data

```python
cat_data=df.select_dtypes('object')
for col in cat_data.columns:
    encoder=LabelEncoder()
    df[col]=encoder.fit_transform(df[col])
```

```python
df.info()
```

```python
df.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(df[df.columns[:-1]],df[df.columns[-1]])
```

```python
lr=LogisticRegression().fit(X_train,y_train)
```

```python
out=lr.predict(X_test)
```

```python
print(classification_report(y_test,out))
```
