```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler,LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.naive_bayes import BernoulliNB,GaussianNB
```

```python
df=pd.read_csv('titanic.csv')
```

```python
df.drop(columns=['PassengerId','Name', 'Ticket','Fare', 'Cabin','Embarked'],inplace=True)
```

```python
df.info()
```

```python
scaler=MinMaxScaler()
df['Age']=scaler.fit_transform(df['Age'].values.reshape(-1,1))
# df['Age']=
```

```python
le=LabelEncoder()
y=le.fit_transform(df['Survived'])
```

```python
df['Sex']=LabelEncoder().fit_transform(df['Sex'])
```

```python
df.Embarked.value_counts()
```

```python
df.head()
```

```python
y=df['Survived'].values
df['Age'].fillna(df['Age'].mean(),inplace=True)
X=df[['Pclass','Sex','Age','SibSp','Parch']].values
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=.1,random_state=42)
```

```python
GB=GradientBoostingClassifier(n_estimators=10)
```

```python
GB.fit(X_train,y_train)
```

```python
y_hat=GB.predict(X_test)
```

```python
print(classification_report(y_test,y_hat))
```

```python
XGB=XGBClassifier()
```

```python
XGB.fit(X_train,y_train)
```

```python
x_y_hat=XGB.predict(X_test)
```

```python
print(classification_report(y_test,x_y_hat))
```

```python
nb=BernoulliNB()
nb.fit(X_train,y_train)
```

```python
nb_y=nb.predict(X_test)
```

```python
print(classification_report(y_test,nb_y))
```
